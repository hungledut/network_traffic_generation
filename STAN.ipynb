```python
import pandas as pd
data = pd.read_csv('UNSW_NB15_testing-set.csv')
data = data.drop(columns = 'id')
data
```

```python
import category_encoders as ce
#Create object for one-hot encoding
encoder=ce.OneHotEncoder(cols=['state','proto','attack_cat','service'],handle_unknown='return_nan',return_df=True,use_cat_names=True)
#Fit and transform Data
data_encoded = encoder.fit_transform(data)
data_encoded
```

```python
import torch
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
dataset1_norm = scaler.fit_transform(data_encoded.values)
dataset1 = torch.Tensor(dataset1_norm)
dataset1
```

```python
dataset1.shape
```

```python
X, y = dataset1[:, :-1], dataset1[:, -1]
```

```python
X = X.unsqueeze(-1)
y = y.unsqueeze(-1)
```

```python
X.shape, y.shape
```

```python
from stannetflow import STANSynthesizer
stan = STANSynthesizer(dim_in=1, dim_window=203)
stan.fit(X, y)
samples = stan.sample(10)
print(samples.shape)
```
